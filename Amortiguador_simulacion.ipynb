```python
#amortiguador simular movimiento
import matplotlib.pyplot as plt
import numpy as np
from scipy.integrate import odeint
%matplotlib inline
```

```python
k = 3.0
m = 1.0
B = 0.5

def armonico(variables, t):
    x, y = variables
    return [y, -k * x / m - B / m * y]

inicial = [0.8,0]

tiempo = np.arange(0,20,0.01)

resultado = odeint(armonico, inicial, tiempo)

xx,yy = resultado.T

plt.plot(tiempo, xx, c = 'r',label = "Posicion")
plt.plot(tiempo, yy, c = 'k',label = "Velocidad")
plt.legend(loc='best', prop={'size':14})
plt.xlabel('tiempo',fontsize=14)
```

```python
from ipywidgets import *

def amortiguado(k = 0, m = 1.0, B = 0.5):

    def armonico(variables, t):
        x, y = variables
        return [y, -k * x / m - B / m * y]

    inicial = [0.8,0.5]

    tiempo = np.arange(0,20,0.01)

    resultado = odeint(armonico, inicial, tiempo)

    xx,yy = resultado.T

    plt.plot(tiempo, xx, c = 'r',label = "Posicion")
    plt.plot(tiempo, yy, c = 'k',label = "Velocidad")
    plt.legend(loc='best', prop={'size':14})
    plt.xlabel('tiempo',fontsize=14)
    plt.show()

interact_manual(amortiguado, k =(0,10,0.1), m= (0,10,0.1), B = (0,10, 0.1));
```

```python
r = 1
def poblacion(x,t):
    return r*x*(1-x)

x0 = 0.05
tt = np.linspace(0,10)
xx = odeint(poblacion,x0,tt)
plt.plot(tt,xx)
plt.xlabel('$t$',fontsize=18)
plt.ylabel('$x$',fontsize=18)
plt.show
```

```python
def Robmovil(x,t):
    R=1
    L=0.5
    vr=0.2
    vl=0.1
    phi=x[2]
    dx=R/2 * (vr + vl) * np.cos(phi)
    dy=R/2 * (vr + vl) * np.sin(phi)
    dphi=(R/L) * (vr-vl)
    return [dx,dy,dphi]

inicial = [0,0,0]
tiempo = np.arange(0,20,0.1)
edos = odeint(Robmovil,inicial,tiempo)
x = edos[:,0]
y = edos[:,1]
plt.plot(x,y,c = 'r', label="holi")
```
